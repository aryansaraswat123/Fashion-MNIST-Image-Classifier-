```python
from __future__ import absolute_import, division, print_function, unicode_literals
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
print(tf.__version__)
```

get the fashion MNIST dataset from teras package - obtain the training set and test set

```python
fashion_mnist = keras.datasets.fashion_mnist
(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()
```

```python
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Sandal', 'Coat', 'Dress', 'Shirt', 'Sneaker', 'Bag', 'Angel Boot']
```

```python
print(train_images.shape)
print(len(train_labels))
print((train_labels))

print(test_images.shape)
print(len(test_labels))
print(test_labels)
```

plot the first figure in the training data

```python
plt.figure()
plt.imshow(train_images[0])
plt.colorbar()
plt.grid(False)
plt.show()
```

normalize the train and test images

```python
train_images = train_images/255.0
test_images = test_images/255.0
```

plot the sample training images

```python
plt.figure(figsize=(10,10))
for i in range(25):
  plt.subplot(5,5,i+1)
  plt.xticks([])
  plt.yticks([])
  plt.grid(False)
  plt.imshow(train_images[i], cmap=plt.cm.binary)
  plt.show()
```

initialize the modle with one hidden layer(no. of neurons = 128)

check the number of parameters in the model

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape = (28,28)),
    keras.layers.Dense(128, activation = 'relu'),
    keras.layers.Dense(10, activation = 'softmax')
])

model.summary()
```

set the parameters for the model

```python
model.compile(optimizer= 'adam', loss = 'sparse_categorical_crossentropy', metrics = ['accuracy'])

```

train the model for 10 epochs

if batch_size = 32, then the number of batches = 60000/32 = 1875

```python
model.fit(train_images, train_labels, epochs = 10, verbose = 1)
```

evaluate the model and evaluatw the accuracy and loss

```python
test_loss, test_acc = model.evaluate(test_images, test_labels)
print("\n Test accuracy",  test_acc)
print("\n Test loss", test_loss)
```

make predictions on Test images

look at the probablity score obtained on predicting the Test_image[0]


```python
prediction = model.predict(test_images)
print(prediction[0])

print('\n predicted value: ', np.argmax(prediction[0]))
print('\n groundtruth value; ' , test_labels[0])
```

function to plot image

```python
def plot_image(i, predictions_array, true_label, img):
  predictions_array, true_label, img = predictions_array[i], true_label[i], img[i]
  plt.grid(False)
  plt.xticks([])
  plt.yticks([])
  plt.imshow(img, cmap=plt.cm.binary)
  predicted_label = np.argmax(predictions_array)
  if predicted_label == true_label:
    color = 'blue'
  else:
    color = 'red'
    plt.xlabel("{} {:2.0f}% ({})".format(class_names[predicted_label], 100*np.max(predictions_array), class_names[true_label]), color = color)
```

```python
def plot_value_array(i, predictions_array, true_label):
    predictions_array, true_label = predictions_array[i], true_label[i]
    plt.grid(False)
    plt.xticks(range(10))
    plt.yticks([])
    thisplot = plt.bar(range(10), predictions_array, color = "#777777")
    plt.ylim([0,1])
    predicted_label = np.argmax(predictions_array)
    thisplot[predicted_label].set_color('red')
    thisplot[true_label].set_color('blue')
```

```python
i=6
plt.figure(figsize=(6,3))
plt.subplot(1,2,1)
plot_image(i, prediction, test_labels, test_images)
plt.subplot(1,2,2)
plot_value_array(i, prediction, test_labels)
plt.show()
```

visualize the 15 images with the prediction probablity scores

```python
num_rows=5
num_cols=3
num_images = num_rows*num_cols
plt.figure(figsize=(2*2*num_cols, 2*num_rows))
for i in range(num_images):
  plt.subplot(num_rows, 2*num_cols, 2*i+1)
  plot_image(i, prediction, test_labels, test_images)
  plt.subplot(num_rows, 2*num_cols, 2*i+2)
  plot_value_array(i, prediction, test_labels)

plt.show()
```
